```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
Raw_DIR= r'D:\mrlEyes_2018_01 (1)\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4]=='0':
            shutil.copy(src=dirpath+'/'+i, dst=r'D:\mrlEyes_2018_01 (1)\Prepared_Data\Closed Eyes')
        
        elif i.split('_')[4]=='1':
            shutil.copy(src=dirpath+'/'+i, dst=r'D:\mrlEyes_2018_01 (1)\Prepared_Data\Opened Eyes')
```
